```python

question = "Should I borrow money to buy Bitcoin?"

high_risk_words = [
    "borrow",
    "loan",
    "bitcoin",
    "crypto",
    "gambling",
    "forex"
]

risk_found = False

for word in high_risk_words:
    if word in question.lower():
        risk_found = True

if risk_found:
    print("⚠️ HIGH RISK WARNING")
    print("Your question contains risky financial decisions.")
    print("Question:", question)
else:
    print("✅ LOW RISK")
    print("Question:", question)
```

```python
question = "How can I create a monthly budget?"

high_risk_words = {
    "borrow": {
        "category": "Debt Risk",
        "reason": "Borrowing money to invest can lead to financial hardship if the investment loses value."
    },
    "loan": {
        "category": "Debt Risk",
        "reason": "Loans increase financial obligations and investment returns are never guaranteed."
    },
    "bitcoin": {
        "category": "Investment Risk",
        "reason": "Bitcoin is highly volatile and its price can change dramatically."
    },
    "crypto": {
        "category": "Investment Risk",
        "reason": "Cryptocurrency investments carry significant market risk."
    },
    "gambling": {
        "category": "Behavioral Risk",
        "reason": "Gambling can result in substantial financial losses."
    },
    "forex": {
        "category": "Trading Risk",
        "reason": "Forex trading is highly leveraged and risky."
    }
}

risk_detected = False

for word in high_risk_words:
    if word in question.lower():
        risk = high_risk_words[word]

        print("⚠️ HIGH RISK WARNING\n")
        print("Category:", risk["category"])
        print("Reason:", risk["reason"])

        print("\nSafer Advice:")
        print("- Invest only money you can afford to lose.")
        print("- Never rely on guaranteed profits.")
        print("- Consider speaking with a licensed financial adviser.")

        risk_detected = True
        break

if not risk_detected:
    print("✅ LOW RISK")
    print("No major financial safety concerns detected.")
```

```python
questions = [
    "Should I borrow money to buy Bitcoin?",
    "How can I create a monthly budget?",
    "Should I take a loan to trade forex?",
    "Is it safe to invest all my salary in crypto?",
    "How do I save money every month?"
]

high_risk_words = {
    "borrow": ("Debt Risk", "Borrowing money to invest can lead to financial hardship."),
    "loan": ("Debt Risk", "Loans increase financial obligations and investment returns are never guaranteed."),
    "bitcoin": ("Investment Risk", "Bitcoin is highly volatile."),
    "crypto": ("Investment Risk", "Cryptocurrency investments carry significant market risk."),
    "forex": ("Trading Risk", "Forex trading is highly leveraged and risky."),
    "all my salary": ("Financial Planning Risk", "Investing all your income leaves no financial safety buffer.")
}

for question in questions:

    print("=" * 60)
    print("Question:")
    print(question)
    print()

    risk_found = False

    for word, details in high_risk_words.items():
        if word in question.lower():

            print("⚠️ HIGH RISK")
            print("Category:", details[0])
            print("Reason:", details[1])

            print("\nSafety Recommendation:")
            print("- Diversify your investments.")
            print("- Avoid decisions based on emotions.")
            print("- Seek professional financial advice.")

            risk_found = True
            break

    if not risk_found:
        print("✅ LOW RISK")
        print("No major financial safety concerns detected.")

print("=" * 60)
print("Analysis Complete")
```
